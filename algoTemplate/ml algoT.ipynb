## Example (Linear Regression)

### A. Hypothesis Function
$$
\Large h_\theta(x) = \theta_0 + \theta_1 x
$$

Where:
- $h_{\theta}(x)$ is the predicted output (hypothesis).
- $\theta_0$ is the y-intercept (bias term).
- $\theta_1$ is the slope (weight for the feature).
- x is the input feature.

For **multivariable linear regression** (with multiple features), the hypothesis function can be written as:

$$
\Large h_\theta(x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n = \theta^T x
$$

Where:
- $\theta_T$ is the transpose of the parameter vector **theta**.
- x is the feature vector.

---

### B. Cost Function (SSE and MSE)

The **cost function** in linear regression measures the error between the predicted values and the actual target values. 

#### B.1 Sum of Squared Errors (SSE)

$$
\Large SSE = \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2
$$

Where:
- m is the number of training examples.
- $h_\theta(x^{(i)})$ is the predicted value for the ith training example.
- $y^{(i)}$ is the actual target value for the ith training example.

#### B.2 Mean Squared Error (MSE)

$$
\Large MSE = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2
$$

---

### C. Gradient Descent Algorithm

The **Gradient Descent** algorithm is used to minimize the cost function by iteratively adjusting the parameters $\theta_0$ and $\theta_1$. 

$$
\Large \theta_j = \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)}
$$

Where:
- alpha is the learning rate.
- $\theta_j$ represents the parameters to be updated.
- $x_j^{(i)}$ is the feature for the ith training example.

#### Gradient Descent Steps:
1. Initialize parameters $\theta_0$, $\theta_1$.
2. Compute the gradient of the cost function with respect to $\theta_0$ and $\theta_1$.
3. Update the parameters using the gradient descent rule (learning rate).
4. Repeat until convergence.
